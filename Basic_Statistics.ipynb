```python
# Defined Function for calculating IQR:
```

```python
def calculate_iqr(df):
    numerical_col=df.select_dtypes(include='number').columns
    for column in numerical_col:
        Q1=df[column].quantile(0.25)
        Q3=df[column].quantile(0.75)
        IQR= Q3 - Q1
        print(f'\nColumn: {column}')
        print(f'Q1: {Q1}')
        print(f'Q3: {Q3}')
        print(f'IQR: {IQR}')
```

```python
calculate_iqr(df)
```

```python
# Define Function for Counting Outliers: 
```

def count_outliers(df):
    numerical_cols=df.select_dtypes(include='number').columns
    for col in numerical_cols:
        Q1=df[col].quantile(0.25)
        Q3=df[col].quantile(0.75)
        IQR= Q3 - Q1
        lower_bound= Q1 - 1.5 * IQR
        upper_bound= Q3 + 1.5 * IQR
        outliers=df[(df[col]<lower_bound) | (df[col]>upper_bound)]
        print(f'\nColumn: {col} -> Outlier Count = {outliers.shape[0]}')

```python
count_outliers(df)
```

## Standardization of Numerical Variables

```python
#identification of numerical variables:
```

```python
numerical_cols=df.select_dtypes(include='number').columns
numerical_cols
```

```python
#Before Standardization - Histograms
```

```python
plt.figure(figsize=(12,10))
df[numerical_cols].hist(bins=20, figsize=(12,10))
plt.suptitle('Before Standardization - Histograms', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#  Standardization using Z-score formula
```

```python
df_standardized = df.copy()
```

```python
df_standardized[numerical_cols]=(df[numerical_cols]-df[numerical_cols].mean()) / df[numerical_cols].std()
```

```python
#  After Standardization: Histograms
```

```python
plt.figure(figsize=(12,10))
df_standardized[numerical_cols].hist(bins=20, figsize=(12,10))
plt.suptitle('After Standardization - Histograms', fontsize=16)
plt.tight_layout()
plt.show()
```
